```python
new_df
```

```python
new_df.sort_values(by='Year', ascending=True, inplace=True)

# Tentukan palet warna yang berbeda untuk setiap tahun
color_palette = plt.get_cmap('tab20', len(new_df))

# Sekarang Anda dapat membuat plot dengan Matplotlib
plt.figure(figsize=(20, 10))  # Anda dapat mengatur ukuran plot
bars = plt.barh(new_df['Year'], new_df['Total'], color=[color_palette(i) for i in np.arange(len(new_df))])
plt.title('Total Data per Year')
plt.xlabel('Total')
plt.ylabel('Year')

# Tambahkan legenda dengan warna untuk setiap tahun
# Tambahkan legenda dengan tahun
for bar, year in zip(bars, new_df['Year']):
    plt.text(bar.get_width() + 5, bar.get_y() + bar.get_height() / 2, str(year), ha='left', va='center')
    
legend_labels = [str(year) for year in new_df['Year']]
legend_handles = [plt.Rectangle((0, 0), 1, 1, color=color_palette(i)) for i in np.arange(len(new_df))]
plt.legend(legend_handles, legend_labels, title='Year', loc='lower right')

plt.show()
```
