## Melihat Penjualan historis dari tahun ke tahun

```python
total_2020 = df[df["Tahun_Mobil"]==2020].groupby("Nama_Mobil")["harga_jual"].sum().sum()
```

```python
total_2019 = df[df["Tahun_Mobil"]==2019].groupby("Nama_Mobil")["harga_jual"].sum().sum()
```

```python
total_2018 = df[df["Tahun_Mobil"]==2018].groupby("Nama_Mobil")["harga_jual"].sum().sum()
```

```python
total_2017 = df[df["Tahun_Mobil"]==2017].groupby("Nama_Mobil")["harga_jual"].sum().sum()
```

```python
total_2015 = df[df["Tahun_Mobil"]==2015].groupby("Nama_Mobil")["harga_jual"].sum().sum()
```

```python
total_2014 = df[df["Tahun_Mobil"]==2014].groupby("Nama_Mobil")["harga_jual"].sum().sum()
```

```python
total_2013 = df[df["Tahun_Mobil"]==2013].groupby("Nama_Mobil")["harga_jual"].sum().sum()
```

```python
total_2012 = df[df["Tahun_Mobil"]==2012].groupby("Nama_Mobil")["harga_jual"].sum().sum()
```

```python
total_2011 = df[df["Tahun_Mobil"]==2011].groupby("Nama_Mobil")["harga_jual"].sum().sum()
```

```python
total_2010 = df[df["Tahun_Mobil"]==2010].groupby("Nama_Mobil")["harga_jual"].sum().sum()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Definisikan data Anda dalam bentuk dictionary atau list of dictionaries
data = [
    {'Year': 2020, 'Total': total_2020},
    {'Year': 2019, 'Total': total_2019},
    {'Year': 2018, 'Total': total_2018},
    {'Year': 2017, 'Total': total_2017},
    {'Year': 2015, 'Total': total_2015},
    {'Year': 2014, 'Total': total_2014},
    {'Year': 2013, 'Total': total_2013},
    {'Year': 2012, 'Total': total_2012},
    {'Year': 2011, 'Total': total_2011},
    {'Year': 2010, 'Total': total_2010},
]

# Buat DataFrame dari data
new_df = pd.DataFrame(data)

# Urutkan DataFrame berdasarkan kolom 'Year' secara ascending
new_df.sort_values(by='Year', ascending=True, inplace=True)

# Sekarang Anda dapat membuat plot dengan Seaborn
plt.figure(figsize=(10, 6))  # Anda dapat mengatur ukuran plot
sns.barplot(x='Total', y='Year', data=new_df, orient='h')  # Gunakan orient='h' untuk bar plot horizontal
plt.title('Total Data per Year')
plt.show()
```
